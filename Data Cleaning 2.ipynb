## 1. Production Activities
### 1.1. For the file (Export of goods and services total and by product (2016).xlsx)

```python
#Import data using pandas
pa_exports=pd.read_excel(r"/Users/edgartome_1/IronHack/IronProjects/Project4/Data_Economic_Activities/Production_Activities/Export of goods and services total and by product (2016).xlsx")
pa_exports
```

```python
#Liste the name of columns
pa_exports.columns
```

```python
#Place the name "Years" on the row that i will have the columns name, then drop the rest of columns and rows
pa_exports.loc[pa_exports.index[6], 'Unnamed: 0'] = "Years"
```

```python
#Delete all rows with data not relevant
pa_exports = pa_exports.drop(labels=range(0, 6), axis=0)
```

```python
#Delete all rows with data not relevant
pa_exports = pa_exports.drop(labels=range(32, 54), axis=0)
```

```python
#Delete all columns with NAN values
pa_exports = pa_exports.dropna(thresh=10, axis=1)
```

```python
#Rename the columns name with the values of the row index 6 
pa_exports.rename(columns=pa_exports.iloc[0], inplace = True)
```

```python
#Rename the columns by code "Col_" plus position
pa_exports.columns = ['Col_' + str(i) if 3 <= i <= 39 else x for i, x in enumerate(pa_exports.columns, 1)]
```

```python
#Reset index
pa_exports = pa_exports.reset_index()
```

```python
#Delete old index
pa_exports = pa_exports.drop(labels=['index'], axis=1)
```

```python
#Delete old columns name
pa_exports = pa_exports.drop(labels=[0], axis=0)
```

```python
#Convert to interger and floats, because on data visualiztion verifcstion that somo tables had the data in objects
pa_exports.apply(pd.to_numeric)
```

```python
#Delete columns Col_39
pa_exports = pa_exports.drop(labels=['Col_39'], axis=1)
```

```python
#Transpose
pa_exports = pa_exports.T

#Rename the columns name with the values of the row index 0 
pa_exports.rename(columns=pa_exports.iloc[0], inplace = True)

pa_exports = pa_exports.drop(labels=['Years'], axis=0)
pa_exports = pa_exports.drop(labels=['Total'], axis=0)


cols=range(1995, 2020)
pa_exports.columns=cols
```

```python
#Export to CSV file
pa_exports.to_csv(r'/Users/edgartome_1/IronHack/IronProjects/Project4/Data_Economic_Activities/Production_Activities/Export_of_goods.csv', index=True)
pa_exports
```
